**map**
- it is one method to convert catagorical value to numerical value.
- take one catagorical column
- get the unique lables first
- make a dictionary with assign a no. to each lable
- for example case status has 2 lables
- 1.cetified
- 2.denied
- assign 0 to certified ,1 to denied
- cret a dictionary lables as keys,no.asvalues only
- {certified:0,denied:1}

```python
d={'Certified':0,'Denied':1}
visa_df['case_status_new']=visa_df['case_status'].map(d)
visa_df
```

```python
#i want to apply the map method for all catagorical columns
#step_1:we need to get unique lbles
#step_3:we need to get a no.=len unique lables
#step_4:we need to creat a dictionary
```

```python
unique_lables=sorted(visa_df['case_status'].unique())
list1=[i for i in range(len(unique_lables))] 
unique_lables,list1
d=dict(zip(unique_lables,list1))
visa_df['case_status']=visa_df['case_status'].map(d)
visa_df
```

```python
for i in cat:
    unique_lables=sorted(visa_df[i].unique())
    list1=[i for i in range(len(unique_lables))]
    d=dict(zip(unique_lables,list1))
    visa_df[i]=visa_df[i].map(d)
visa_df
```

**advantages**

- using one hot encoder we will get new columns
- the new column value at a time only one will be true ,another will be false.
- which means new columns are independent each other
- independent means no relation
- means no **multi colineraity**
- multi colinearity is animportant property in machine learning
- **ml models expect the column should be independent each other**
- so one hot encoder use the nomulticolineartity

**disadvantages**
- it will increase many columns
- inthe aboveex.total column becomes 25510
- this is happendbecause of case id
- in realtime we will drop the id columns
- eventhough if we drop stillwe are getting 30 columns
- more columns means ,more time to process
- we required more infrastructure(cpu-gpu-tpu)
- this is clled as **curse of dimensionality**
