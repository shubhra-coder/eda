- using d types

- whenever u want to creat any list from series

- 1st convert it into dictionary

- when u convert it becomes a key value pair

- iterate through dictionary

- if the value==obj type,then keep that corresponding key in one list

- else keep that variable in other list

```python
dict(visa_df.dtypes)
```

```python
dict(visa_df.dtypes).items()
```

```python
data_types=visa_df.dtypes
dict(data_types)
for i,j in dict(data_types).items():
    print(i,j)
```

```python
data_types=visa_df.dtypes
dict(data_types)
l1=[]
l2=[]
for i,j in dict(data_types).items():
    if j=="object":
        l1.append(i)
    else:
        l2.append(i)
print(l1)   
print(l2)
```

```python
cat_list=[i for i,j in dict(data_types).items() if j=="object"]
num_list=[i for i,j in dict(data_types).items() if j!="object"]
print(cat_list)
print(num_list)
```
